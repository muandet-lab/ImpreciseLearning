```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
lambdas = np.array([0.0, 0.1, 0.3, 0.5, 0.7, 0.9, 0.999])
data = {
    "GT": {
        "mean": np.array([0.209, 0.230, 0.267, 0.297, 0.316, 0.325, 0.343]),
        "std": np.array([9.38e-03, 1.10e-06, 2.45e-07, 2.62e-08, 1.34e-08, 4.33e-03, 4.33e-03])
    },
    "Median": {
        "mean": np.array([0.345, 0.382, 0.489, 0.681, 1.039, 1.231, 1.297]),
        "std": np.array([4.23e-02, 4.65e-02, 6.10e-02, 8.79e-02, 1.26e-02, 1.40e-01, 1.51e-01])
    },
    "Avg": {
        "mean": np.array([0.209, 0.230, 0.271, 0.327, 0.455, 0.577, 0.614]),
        "std": np.array([3.60e-03, 3.66e-03, 4.75e-03, 8.30e-03, 1.74e-02, 2.10e-02, 2.57e-02])
    },
    "Worst": {
        "mean": np.array([0.247, 0.271, 0.310, 0.321, 0.326, 0.333, 0.343]),
        "std": np.array([6.78e-03, 6.93e-03, 7.44e-03, 8.00e-03, 8.18e-03, 7.17e-03, 5.91e-03])
    },
    "Uniform": {
        "mean": np.array([0.235, 0.259, 0.295, 0.308, 0.319, 0.341, 0.363]),
        "std": np.array([5.17e-03, 5.69e-03, 7.19e-03, 7.23e-03, 5.06e-03, 1.14e-02, 1.34e-02])
    },
    "Unif(0,1)h": {
        "mean": np.array([0.209, 0.230, 0.269, 0.296, 0.318, 0.327, 0.351]),
        "std": np.array([6.00e-04, 1.07e-03, 1.38e-03, 5.22e-05, 1.10e-03, 2.92e-03, 1.07e-02])
    }
}
custom_legend_names = {
    "GT": "GT ($\lambda$ known)",
    "Reg": "Two-stage Reg.",
    "Beta(0.1,0.2)": "Beta(0.1,0.2)",
    "Chebyshev": "Chebyshev",
    "Median": "Median",
    "Avg": "$\lambda=0$",
    "Worst": "$\lambda$->$1$",
    "Uniform": "$Unif[0,1]$ $f(x)$",
    "Unif(0,1)h": "$Unif[0,1]$ $h(x,\lambda)$"
}

# Different color for each learner
colors = {
    "GT": "blue", "Median": "brown", 
    "Avg": "green", "Worst": "purple", 
    "Uniform": "orange", "Unif(0,1)h": "red"
}

# Create the plot
plt.figure(figsize=(4, 3))
for learner, color in colors.items():
    plt.plot(lambdas, data[learner]["mean"], label=custom_legend_names[learner], color=color)
    plt.fill_between(lambdas, 
                     data[learner]["mean"] - data[learner]["std"], 
                     data[learner]["mean"] + data[learner]["std"], 
                     color=color, alpha=0.2)

# Limit y-axis to exclude outlier values
plt.ylim(0.2, 0.4)  # Set y-axis limit

# Labels and title
plt.xlabel('$\lambda$', fontsize='large')
plt.ylabel('$CVaR_\lambda(\mathbf{\mathcal{R}})$',fontsize='large')
plt.title('Precise vs Imprecise Learners')
plt.legend(fontsize=7)
plt.grid(True)
plt.savefig('precisevsimprecise.pdf', format='pdf', bbox_inches='tight')
# Show the plot
plt.show()
```
