```python
import numpy as np

risks = np.random.beta(2, 3, size=25)
true_samples = np.random.beta(2, 3, size=100000)
points = risks[np.random.randint(0,len(risks),100000)]
samples =  np.array([point + 0.1*np.random.normal(0,1) for point in points])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(risks, label="observations")
sns.distplot(true_samples, label="true")
sns.distplot(samples, label="estimate")
plt.legend()
plt.show()
```

```python
alpha = 0.99
risk_quantile = np.quantile(risks, alpha)
sample_quantile = np.quantile(samples, alpha)
true_quantile = np.quantile(true_samples, alpha)

risk_cvar = risks[risks >= risk_quantile].mean()
sample_cvar = samples[samples >= sample_quantile].mean()
true_cvar = true_samples[true_samples >= true_quantile].mean()

print(risk_cvar)
print(sample_cvar)
print(true_cvar)
```
